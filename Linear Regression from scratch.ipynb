Formulae's for MSE, Partial derivative of coeficient(slope) and intercept(b).

```python
import math
import numpy as np
```

```python
def grad_descent(x,y):
    m_cur = b_cur = 0
    iterations = 100
    learning_rate = 0.03
    n = len(x) #Assuming len of both are same(it should be same)
    for i in range(iterations):
        y_pred = m_cur * x + b_cur
        cost = 1/n * sum(y-y_pred)**2
        md = -(2/n) * sum(x*(y-y_pred))
        bd = -(2/n) * sum((y-y_pred))
        m_cur = m_cur - learning_rate * md
        b_cur = b_cur - learning_rate * bd
        print(m_cur, b_cur, cost, i)
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

grad_descent(x, y)
```
